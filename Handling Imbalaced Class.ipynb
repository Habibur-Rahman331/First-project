#Encodes the Dataset

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df.columns
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in df. columns:
    if is_numeric_dtype(df[column]):
        continue


    else:
        df[column] = label.fit_transform(df[column])
        
```
